```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('spectrum-timing.csv', index_col=0)
df.head()
```

```python
sns.scatterplot(data=df, x='L', y='gs_time')
plt.xlabel('System size')
plt.ylabel('Time (s)')
plt.title('Ground state search time')
```

```python
sns.scatterplot(data=df, x='L', y='sf_time')
plt.xlabel('System size')
plt.ylabel('Time (s)')
plt.title('Spectral function computation time ($\langle S^z(t) S^z(0) \\rangle_0$)')
```
